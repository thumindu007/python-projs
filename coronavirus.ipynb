```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
height = np.array([147096,2717292,75807,1247975,566507,177582,2034602,855855])
x = ['ACT', 'NSW', 'NT', 'QLD', 'SA', 'TAS', 'VIC', 'WA']

plt.bar(x, height)
plt.show()
```
